```python
import numpy as np, time
from sklearn.naive_bayes import CategoricalNB
print("imports ok")
```

```python
# batch size comparison
X = np.random.randint(0, 100, (5_000_000, 7)).astype(np.int32)
y = np.random.randint(0, 2, 5_000_000).astype(bool)
for B in [100000, 500000, 1000000]:
    clf = CategoricalNB(alpha=1.0, min_categories=100)
    t0 = time.time()
    for i in range(0, len(X), B):
        clf.partial_fit(X[i:i+B], y[i:i+B], classes=[0, 1])
    print(f"batch={B//1000}k: {time.time()-t0:.2f}s")
```

```python
# fit vs partial_fit equivalence
X2 = np.random.randint(0, 100, (500_000, 7)).astype(np.int32)
y2 = np.random.randint(0, 2, 500_000).astype(bool)
clf1 = CategoricalNB(alpha=1.0, min_categories=100)
clf1.fit(X2, y2)
clf2 = CategoricalNB(alpha=1.0, min_categories=100)
for i in range(0, len(X2), 100000):
    clf2.partial_fit(X2[i:i+100000], y2[i:i+100000], classes=[0, 1])
err = np.abs(clf1.feature_log_prob_[0] - clf2.feature_log_prob_[0]).max()
print(f"max diff fit vs partial_fit: {err:.2e}")
```

```python
# memory: int32 vs float64
for dtype, name in [(np.int32, "int32"), (np.float32, "float32"), (np.float64, "float64")]:
    sz = np.ones((5_000_000, 7), dtype=dtype).nbytes / 1024**2
    print(f"{name}: {sz:.0f} MB")
```
