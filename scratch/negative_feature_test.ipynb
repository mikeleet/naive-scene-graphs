```python
import numpy as np
from sklearn.naive_bayes import CategoricalNB

DATA = "data"
```

```python
# only feature 4 (rel_area) can be negative [originally -9..9]
# solution: shift +9 -> [0..18]
print("feature 4 (rel_area): -9..9 -> shifted +9 -> 0..18")
```

```python
# sklearn rejects negative values
np.random.seed(1)
X_bad = np.column_stack([
    np.random.randint(0, 150, 1000),
    np.random.randint(0, 150, 1000),
    np.random.randint(0, 8, 1000),
    np.random.randint(0, 8, 1000),
    np.random.randint(-9, 10, 1000),  # negative!
    np.random.randint(0, 10, 1000),
    np.random.randint(0, 10, 1000),
])
y = np.random.randint(0, 3, 1000)
try:
    CategoricalNB(min_categories=[150,150,8,8,19,10,10]).fit(X_bad, y)
except ValueError as e:
    print(f"negatives fail as expected: {e}")

X_good = X_bad.copy()
X_good[:, 4] += 9
CategoricalNB(min_categories=[150,150,8,8,19,10,10]).fit(X_good, y)
print("shifted works")
```

```python
# verify on actual data
X = np.load(f"{DATA}/preprocessed/train_X.npy")
names = ["head_cls","tail_cls","topology","angle","rel_area","head_area","tail_area"]
for i, name in enumerate(names):
    vals = X[:, i]
    print(f"{name:15s}: min={vals.min():4d} max={vals.max():4d} unique={len(np.unique(vals))}")
print(f"\nall non-negative: {X.min() >= 0}")
```
